```python
import numpy as np
from numpy import linalg as LA
```

```python
arr_1 = np.array([[14,-7],[9,12]])
arr_2 = np.array([[6],[-9]])

print(arr_1)
print(arr_2)



# arr_5 = np.dot(LA.inv(arr_1),arr_2))
```

```python
arr_3 = np.dot(arr_1,arr_2)
print(arr_3)
```

```python
arr_4 = LA.matrix_power(arr_1,4)
print(arr_4)
```

```python
arr_5 = LA.inv(arr_1)
print(arr_5)
```

```python
# np.linalg.multi_dot()

arr_5 = np.array([[1, 2],
                  [3, 4]])

arr_6 = np.array([[2, 4],
                  [6, 9]])

ans = np.dot(arr_5,arr_6)


arr_7 = np.array([[5, 6],
                  [7, 8]])

print(arr_5)
print(arr_6)

print(ans)

print(arr_7)
```

```python
result = LA.multi_dot((arr_5,arr_6,arr_7))

print(result)
```
