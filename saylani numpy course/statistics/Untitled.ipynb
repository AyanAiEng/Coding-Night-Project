```python
import numpy as np
import math
```

```python
in_invest = 1000
roc = np.array([1,-0.5])
roc = roc + 1
product = np.prod(roc)
gmean = math.pow(product,1/len(roc)) -1
print(in_invest*(1-gmean))
```
