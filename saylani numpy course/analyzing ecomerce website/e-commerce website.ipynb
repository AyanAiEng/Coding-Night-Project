Find which product is top sailing product on the basis of repeated orders.

```python
product_id = order[:,2]
# print(product_id)
best_products,count = np.unique(product_id,return_counts = True)
top_product = np.max(count)
print(top_product)
is_top = (count == top_product)
product_count = count[is_top]
best_products = best_products[is_top]
print(best_products,count)
print(is_top)
print(product_count)
print(best_products) 
print(product_count)
top_products = np.vstack((best_products, product_count)).T
print(top_products)
print(f"The top selling products are {best_products} with the count of{product_count}")
```

Find which customer should get more discount as he/she keep visiting my store more than 5 tines.

```python
order = np.loadtxt("order.csv",delimiter = ',',dtype = str)
np.set_printoptions(threshold = np.inf)
order = np.delete(order,0,axis = 0)
customer_id = order[:,1]
order_id = order[:,0]

unique_element,count = np.unique(customer_id,return_counts = True) 

mask = count > 5

discount_customers = unique_element[mask]
order_count = count[mask]


sorted_indices = np.argsort(order_count)
sorted_customers = discount_customers[sorted_indices]
sorted_count = order_count[sorted_indices]

answer = np.column_stack((sorted_customers,sorted_count))

print(mask)
print(unique_element,count)
print(discount_customers)
print(order_count)
print(sorted_indices)
print(sorted_customers)
print(sorted_count)
print(answer)
print(f"These are some of the customers to which discount should be given {sorted_customers}")
```
